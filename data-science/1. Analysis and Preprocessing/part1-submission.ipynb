## 1. Find all variables and understand them

```python
#step 1a Load chosen datasets from the datasets folder into your notebook and display the first ten instances.
import pandas as pd
from matplotlib import pyplot as plt

store_sales = pd.read_csv('store_sales.csv', encoding='latin1')
#remove rows with missing data
store_sales = store_sales.dropna(axis=0,how='any')

#first 10 instances
print(store_sales.head(10))
```

```python
#step 1b,c Provide key statistical measures such as mean and standard deviation without hardcoding them, Visualize numerical columns through appropriate plots, e.g. bar plots, histograms, box and whisker plots, etc.

#key statistical measures for numeric variables 

def show_numeric_measures(df, column_name):
    min_val = df.min()
    max_val = df.max()
    mean_val = df.mean()
    med_val = df.median()
    mod_val = df.mode()[0]
    std_val = df.std()
    print('Minimum:{:.2f}\nMean:{:.2f}\nMedian:{:.2f}\nMode:{:.2f}\nMaximum:{:.2f}\n'.format(min_val, mean_val, med_val,mod_val,max_val))
    
    # Create a figure for 2 subplots (2 rows, 1 column)
    fig, ax = plt.subplots(2, 1, figsize = (10,4))
        # Plot the histogram   
    ax[0].hist(df)
    ax[0].set_ylabel('Frequency')

    # Add lines for the mean, median, and mode
    ax[0].axvline(x=min_val, color = 'gray', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mean_val, color = 'cyan', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=med_val, color = 'red', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=mod_val, color = 'yellow', linestyle='dashed', linewidth = 2)
    ax[0].axvline(x=max_val, color = 'gray', linestyle='dashed', linewidth = 2)

    # Plot the boxplot   
    ax[1].boxplot(df, vert=False)
    ax[1].set_xlabel('Value')

    # Title
    fig.suptitle(f"Data Distribution of {column_name}")

    # Display plot
    plt.tight_layout()
    plt.show()
    
def show_categorial_measures(df, column_name):
    unique_values = df[column_name].unique()
    most_frequent = unique_values.value_counts().idxmax()
    print(f'Data for {column_name}:\n Number of unique values: {unique_values}\n Most frequent value: {most_frequent}')
    
def show_datetime_measures(df, column_name):
    min_date = df.index.min()
    max_date = df.index.max()
    mean_date = df.index.mean()
    med_date = df.index.median()
    mod_date = df.index.mode()[0]
    
#order id stats
print(f"Order ID Mode: {store_sales.mode()['Order ID'].iloc[0]}")


for column in store_sales.select_dtypes(include='number').columns:
    #values where 
    if column not in ['Row ID', 'Order ID']:
        show_numeric_measures(store_sales[column],column)

store_sales['Order Date'] = pd.to_datetime(store_sales['Order Date'], errors='coerce')
store_sales['Ship Date'] = pd.to_datetime(store_sales['Ship Date'], errors='coerce')

show_datetime_measures(store_sales, 'Order Date')
show_datetime_measures(store_sales, 'Ship Date')
    
```
